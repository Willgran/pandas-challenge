```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Finding Total Schools with Unique 
Total_Schools = len(school_data_complete.school_name.unique())
#Finding Total Students with Count
Total_Students = school_data_complete.student_name.count()
#Calculating Budget by adding each unique value
Total_Budget = sum(school_data_complete.budget.unique())
#Calculating Math Average by using mean
Math_Average = school_data_complete.math_score.mean()
#Calculating Reading Average by using mean
Reading_Average = school_data_complete.reading_score.mean()
#Calculating Math Passing scores by using scores over 70 and making it a percentage
District_Passing_Math = ((school_data_complete[school_data_complete['math_score'] >= 70 ].math_score.count())/ Total_Students)*100
#Calculating Reading Passing scores by using scores over 70 and making it a percentage
District_Passing_Reading = ((school_data_complete[school_data_complete['reading_score']>= 70 ].math_score.count())/ Total_Students)*100
#Overall Passing Fuction
Overall_Passing = ((District_Passing_Math + District_Passing_Reading)/2)

#Creating a DataFrame to hold and display the values in a table
District_Complete = pd.DataFrame({"Total Schools": [Total_Schools],
                                  "Total Students": [Total_Students],
                                  "Total Budget": [Total_Budget],
                                  "Average Math Score": [Math_Average],
                                  "Average Reading Score": [Reading_Average],
                                  "% Passing Math": [District_Passing_Math],
                                  "% Passing Reading": [District_Passing_Reading],
                                  "% Overall Passing Rate":[Overall_Passing]})
#Cleaning the Data using .map & .format
District_Complete["Total Students"] = District_Complete["Total Students"].map("{:,.0f}".format)
District_Complete["Total Budget"] = District_Complete["Total Budget"].map("${:,.2f}".format)
District_Complete["Average Math Score"] = District_Complete["Average Math Score"].map("{:,.2f}".format)
District_Complete["Average Reading Score"] = District_Complete["Average Reading Score"].map("{:,.2f}".format)
District_Complete["% Passing Math"] = District_Complete["% Passing Math"].map("{:,.2f}%".format)
District_Complete["% Passing Reading"] = District_Complete["% Passing Reading"].map("{:,.2f}%".format)
District_Complete["% Overall Passing Rate"] = District_Complete["% Overall Passing Rate"].map("{:,.2f}%".format)

#Print District Summary
District_Complete
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Creating varaibles to hold school data values, using "school_name" to start each row 

#Creating Types using an index using a key word
School_Types = school_data.set_index(["school_name"])["type"]
#Creating Total Student using Count
School_Total_Students = school_data_complete["school_name"].value_counts()
#Creating a Budget using mean
School_Budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
#Creating Per Student dividing budget by total
Budget_Per_Student = School_Budget/School_Total_Students
#Calculating Average Math using mean
School_Math_Score_Avg = school_data_complete.groupby(["school_name"]).mean()["math_score"]
#Calculating Average Reading using mean
School_Reading_Score_Avg = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
#Calculating Passing Math Percent over 70%
School_Passing_Math = school_data_complete[school_data_complete["math_score"]>= 70]
Percent_Passing_Math = (School_Passing_Math.groupby(["school_name"]).count()["student_name"]/School_Total_Students)*100
#Calculating Passing Reading Percent over 70%
School_Passing_Reading = school_data_complete[school_data_complete["reading_score"]>= 70]
Percent_Passing_Reading = (School_Passing_Reading.groupby(["school_name"]).count()["student_name"]/School_Total_Students)*100
#Calculating Overall Passing
School_Overall_Passing_Percent = (Percent_Passing_Math + Percent_Passing_Reading)/2

#Creating a DataFrame to display the values in a table
School_Summary = pd.DataFrame({"School Type": School_Types,
                                "Total Students": School_Total_Students,
                                "Total School Budget": School_Budget,
                                "Per Student Budget": Budget_Per_Student,
                                "Average Math Score": School_Math_Score_Avg,
                                "Average Reading Score": School_Reading_Score_Avg,
                                "% Passing Math": Percent_Passing_Math,
                                "% Passing Reading": Percent_Passing_Reading,
                                "% Overall Passing Rate": School_Overall_Passing_Percent})
#Creating the Column headers
School_Summary = School_Summary[["School Type","Total Students","Total School Budget", "Per Student Budget","Average Math Score",
                                        "Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing Rate"]]
#Cleaning the data using .map & .format
School_Summary["Total Students"] = School_Summary["Total Students"].map("{:,.0f}".format)
School_Summary["Total School Budget"] = School_Summary["Total School Budget"].map("${:,.2f}".format)
School_Summary["Per Student Budget"] = School_Summary["Per Student Budget"].map("${:,.2f}".format)
School_Summary["Average Math Score"] = School_Summary["Average Math Score"].map("{:,.2f}".format)
School_Summary["Average Reading Score"] = School_Summary["Average Reading Score"].map("{:,.2f}".format)
School_Summary["% Passing Math"] = School_Summary["% Passing Math"].map("{:,.2f}%".format)
School_Summary["% Passing Reading"] = School_Summary["% Passing Reading"].map("{:,.2f}%".format)
School_Summary["% Overall Passing Rate"] = School_Summary["% Overall Passing Rate"].map("{:,.2f}%".format)

#Print School Summary
School_Summary
```
